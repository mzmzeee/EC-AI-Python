## 4. Operators

### Short-Circuit Evaluation

Logical operators `and` and `or` use **short-circuit evaluation**: the second operand is only evaluated if necessary.

- `and`: If the left side is `False`, the whole expression is `False` — the right side is **not evaluated**.
- `or`: If the left side is `True`, the whole expression is `True` — the right side is **not evaluated**.

This is useful for safety checks: `if x and x > 0` avoids a `TypeError` when `x` is `None`.

```python
# Short-circuit examples
# 'and' stops early if left is False
print(0 and 999)          # 0 (left is falsy, returned immediately)
print(5 and 999)          # 999 (left is truthy, evaluates right)

# 'or' stops early if left is True
print(5 or 999)           # 5 (left is truthy, returned immediately)
print(0 or 999)           # 999 (left is falsy, evaluates right)

# Safety pattern: check existence before using
name = None
if name and len(name) > 3:
    print("Long name")
else:
    print("No name or short name")  # This prints because name is None

name = "Alice"
if name and len(name) > 3:
    print("Long name:", name)       # This prints because both conditions pass
```

## 6. For Loops

A `for` loop iterates over any iterable object (lists, strings, ranges). The `range(n)` function generates integers from 0 to n-1; `range(start, stop, step)` provides finer control. The `enumerate()` function returns index-value pairs, eliminating the need for a manual counter.

```python
models = ["ResNet", "VGG16", "EfficientNet"]
for i, model in enumerate(models, start=1):
    print(f"{i}. {model}")
print("Iteration complete.")
```

## 7. While Loops

A `while` loop executes its block as long as a condition remains `True`. The `break` statement terminates the loop immediately; `continue` skips the remaining statements in the current iteration and proceeds to the next. Infinite loops may occur if the condition never becomes false.

```python
count = 5
while count > 0:
    if count == 3:
        count -= 1
        continue                   # skip printing 3
    print(count)
    count -= 1
print("Countdown finished.")
```

## 8. Indexing and Slicing

Sequences (strings, lists, tuples) support zero-based indexing. A single element is accessed via `seq[index]`; negative indices count from the end. Slicing uses the syntax `seq[start:stop:step]`, where `start` is inclusive, `stop` is exclusive, and `step` defaults to 1.

```python
text = "Python"
print(text[0])       # P
print(text[-1])      # n
print(text[0:3])     # Pyt
print(text[::2])     # Pto  (every second character)
```

## 9. Lists

A list is an ordered, mutable collection defined with square brackets `[]`. ## 10. Tuples

Tuples are preferred for fixed data such as image dimensions or hyperparameter configurations. Unpacking assigns tuple elements to individual variables in a single statement.

## 11. Sets

A set is an unordered collection of unique elements, created with `{}` or `set()`. Sets support mathematical operations: `|` (union), `&` (intersection), `-` (difference), and `^` (symmetric difference). Duplicate values are automatically discarded, making sets useful for deduplication.

```python
group_a = {"Omar", "Lina", "Fatima"}
group_b = {"Lina", "Yusuf", "Fatima"}
print("Both:", group_a & group_b)            # intersection
print("A only:", group_a - group_b)           # difference
print("All:", group_a | group_b)              # union
```

## 12. Dictionaries

A dictionary maps keys to values using `{key: value}` syntax. Keys must be immutable (strings, numbers, tuples). Access values with `dict[key]` or the safer `.get(key, default)`. Methods `.keys()`, `.values()`, and `.items()` enable iteration over the dictionary contents.

```python
config = {"lr": 0.001, "batch_size": 32, "epochs": 20}
print("Learning rate:", config["lr"])
print("Dropout:", config.get("dropout", "not specified"))
for key, value in config.items():
    print(f"  {key}: {value}")
```

## 14. Functions

Functions are defined with the `def` keyword, followed by the function name, parameters in parentheses, and a colon. The `return` statement sends a value back to the caller. Default parameter values make arguments optional. A docstring (triple-quoted string immediately after the function header) documents the function's purpose.

```python
def classify(accuracy, threshold=0.90):
    """Return 'Pass' if accuracy meets or exceeds threshold, else 'Fail'."""
    if accuracy >= threshold:
        return "Pass"
    return "Fail"

print(classify(0.95))
print(classify(0.85, threshold=0.80))  # override default
```

### Variable-Length Arguments: `*args` and `**kwargs`

Sometimes you don't know how many arguments a function will receive:

- `*args` collects extra **positional** arguments into a **tuple**.
- `**kwargs` collects extra **keyword** arguments into a **dictionary**.

These are essential when calling library functions that accept many optional parameters.

```python
# *args example: accept any number of numbers
def average_all(*numbers):
    """Return the average of all provided numbers."""
    return sum(numbers) / len(numbers) if numbers else 0

print(average_all(10, 20, 30))      # 20.0
print(average_all(5, 15))           # 10.0

# **kwargs example: accept any keyword arguments
def describe_model(name, **kwargs):
    """Print model info with any extra keyword attributes."""
    print(f"Model: {name}")
    for key, value in kwargs.items():
        print(f"  {key}: {value}")

describe_model("ResNet", layers=50, accuracy=0.94, dataset="ImageNet")

# Combining both
def flexible(*args, **kwargs):
    print("Positional args:", args)
    print("Keyword args:", kwargs)

flexible(1, 2, 3, color="red", size=10)
```

## 15. Scope

Variables defined inside a function are local to that function and are destroyed when the function returns. Variables defined at the module level are global. Python resolves names using the LEGB rule: Local, Enclosing (nonlocal), Global, Built-in. The `global` keyword allows a function to modify a global variable, though this practice should be used sparingly.

```python
counter = 0                 # global

def increment():
    global counter
    local_val = 99           # local - inaccessible outside
    counter += 1
    return local_val

print("Before:", counter)
increment()
print("After:", counter)    # global was modified
# print(local_val)           # NameError
```

## 16. List Comprehensions

A list comprehension builds a new list by evaluating an expression for each item in an iterable, optionally filtering with `if`. The syntax is `[expression for item in iterable if condition]`. Dictionary comprehensions (`{k: v for ...}`) and set comprehensions (`{expr for ...}`) follow the same pattern.

```python
values = [10, 15, 20, 25, 30]
doubled = [v * 2 for v in values if v > 15]
print("Filtered and doubled:", doubled)

# Dict comprehension: square each key
squares = {v: v ** 2 for v in range(1, 5)}
print("Squares dict:", squares)
```

## 17. Lambda Functions

A lambda is an anonymous, single-expression function with the syntax `lambda arguments: expression`. Lambdas are commonly used with `map()` (applies a function to each element) and `filter()` (keeps elements for which the function returns `True`). The return value of `map()` and `filter()` is an iterator; wrap with `list()` to materialise.

```python
numbers = [1, 2, 3, 4, 5]
squared = list(map(lambda x: x ** 2, numbers))
evens = list(filter(lambda x: x % 2 == 0, numbers))
print("Squared:", squared)
print("Evens:", evens)
```

## 19. zip() Function

The `zip()` function accepts two or more iterables and returns an iterator of tuples, where each tuple pairs the elements at corresponding positions. When the input iterables differ in length, `zip()` stops at the shortest. Wrapping the result with `list()` or `tuple()` materialises the iterator for display.


```python
names = ["Omar", "Lina", "Yusuf"]
ages  = [21, 23, 20]
paired = list(zip(names, ages))
print("Name-Age pairs:", paired)

# zip with three iterables
ids = [101, 102, 103]
result = list(zip(ids, names, ages))
print("ID-Name-Age:", result)
```

## 20. sort() with key Parameter

The `sort()` method sorts a list in place. The `key` parameter accepts a function that extracts a comparison value from each element; the list is then sorted according to those extracted values. A `lambda` is commonly used for concise, one-off key functions. The `reverse=True` argument sorts in descending order.


```python
pairs = [("ResNet", 50), ("VGG16", 16), ("EfficientNet", 7)]
pairs.sort(key=lambda x: x[1])          # sort by second element (depth)
print("Sorted by depth:", pairs)

# Sort by first element (name), descending
pairs.sort(key=lambda x: x[0], reverse=True)
print("Sorted by name desc:", pairs)
```
