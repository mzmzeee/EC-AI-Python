## Problem 1: Student Class
Write a program that defines a `Student` class with the following attributes: `name` (str), `age` (int), and `gpa` (float). The `__init__` method should initialize all three attributes. Add a method `is_honors()` that returns `True` if the student's GPA is greater than or equal to 3.5 and `False` otherwise.

Example:
```
s1 = Student("Alice", 20, 3.8)
print(s1.is_honors())  # True

s2 = Student("Bob", 22, 3.2)
print(s2.is_honors())  # False
```

````python
# Write your solution here
````

````python
# SOLUTION
class Student:
    def __init__(self, name, age, gpa):
        self.name = name
        self.age = age
        self.gpa = gpa

    def is_honors(self):
        return self.gpa >= 3.5

# Test
s1 = Student("Alice", 20, 3.8)
s2 = Student("Bob", 22, 3.2)
print(s1.is_honors())  # True
print(s2.is_honors())  # False
````

## Problem 2: BankAccount Class
Create a `BankAccount` class with a private attribute `__balance` initialized to 0. Provide three methods:
- `deposit(amount)`: adds `amount` to the balance if `amount` is positive.
- `withdraw(amount)`: subtracts `amount` from the balance if sufficient funds exist; otherwise, prints an error message.
- `get_balance()`: returns the current balance.

Example:
```
acc = BankAccount()
acc.deposit(100)
acc.withdraw(30)
print(acc.get_balance())   # 70
acc.withdraw(100)          # prints: Insufficient funds.
```

````python
# Write your solution here
````

````python
# SOLUTION
class BankAccount:
    def __init__(self):
        self.__balance = 0

    def deposit(self, amount):
        if amount > 0:
            self.__balance += amount

    def withdraw(self, amount):
        if amount <= self.__balance:
            self.__balance -= amount
        else:
            print("Insufficient funds.")

    def get_balance(self):
        return self.__balance

# Test
acc = BankAccount()
acc.deposit(100)
acc.withdraw(30)
print(acc.get_balance())   # 70
acc.withdraw(100)          # Insufficient funds.
````

## Problem 3: GradStudent (Inheritance)
Define a `GradStudent` class that inherits from `Student` (from Problem 1). It should accept an additional parameter `thesis_topic` (str) in the constructor and pass `name`, `age`, and `gpa` to the parent class. Add a method `research()` that returns a string of the form `"<name> is researching <thesis_topic>."`.

Example:
```
gs = GradStudent("Charlie", 25, 3.9, "Quantum Computing")
print(gs.is_honors())   # True (inherited)
print(gs.research())    # Charlie is researching Quantum Computing.
```

````python
# Write your solution here
````

````python
# SOLUTION
class Student:
    def __init__(self, name, age, gpa):
        self.name = name
        self.age = age
        self.gpa = gpa

    def is_honors(self):
        return self.gpa >= 3.5


class GradStudent(Student):
    def __init__(self, name, age, gpa, thesis_topic):
        super().__init__(name, age, gpa)
        self.thesis_topic = thesis_topic

    def research(self):
        return f"{self.name} is researching {self.thesis_topic}."

# Test
gs = GradStudent("Charlie", 25, 3.9, "Quantum Computing")
print(gs.is_honors())   # True
print(gs.research())    # Charlie is researching Quantum Computing.
````

## Problem 4: File Analysis
Write a program that reads a text file and counts the number of lines, words, and characters (including whitespace) in the file. Print the results in the format: `"Lines: <N>, Words: <M>, Characters: <C>"`.

Assume the file `sample.txt` exists with the following content:
```
Hello world
Python is fun
```
Expected output: `Lines: 2, Words: 5, Characters: 27`

````python
# Write your solution here
````

````python
# SOLUTION
# Create sample file first
with open("sample.txt", "w") as f:
    f.write("Hello world\nPython is fun")

with open("sample.txt", "r") as f:
    content = f.read()

lines = content.splitlines()
num_lines = len(lines)
num_words = len(content.split())
num_chars = len(content)

print(f"Lines: {num_lines}, Words: {num_words}, Characters: {num_chars}")
````

## Problem 5: Number File Sum
Write a program that performs two tasks:
1. Write a list of numbers `[10, 20, 30, 40, 50]` to a file named `numbers.txt`, one number per line.
2. Read the numbers back from the file, convert each line to an integer, and compute (and print) their sum.

Expected output: `Sum: 150`

````python
# Write your solution here
````

````python
# SOLUTION
numbers = [10, 20, 30, 40, 50]

# Write to file
with open("numbers.txt", "w") as f:
    for num in numbers:
        f.write(str(num) + "\n")

# Read back and sum
total = 0
with open("numbers.txt", "r") as f:
    for line in f:
        total += int(line.strip())

print(f"Sum: {total}")
````

## Problem 6: Safe Division
Write a function `safe_divide(a, b)` that returns the result of `a / b`. Use exception handling to catch `ZeroDivisionError` and `TypeError`. In each case, print an appropriate error message and return `None`.

Example:
```
print(safe_divide(10, 2))   # 5.0
print(safe_divide(10, 0))   # Error: Cannot divide by zero. (returns None)
print(safe_divide(10, "2")) # Error: Both arguments must be numbers. (returns None)
```

````python
# Write your solution here
````

````python
# SOLUTION
def safe_divide(a, b):
    try:
        return a / b
    except ZeroDivisionError:
        print("Error: Cannot divide by zero.")
        return None
    except TypeError:
        print("Error: Both arguments must be numbers.")
        return None

# Test
print(safe_divide(10, 2))   # 5.0
print(safe_divide(10, 0))   # Error message and None
print(safe_divide(10, "2")) # Error message and None
````

## Problem 7: NumPy Random Matrix
Using NumPy, create a 5x5 matrix of random integers in the range 1 to 100 (inclusive). Print the matrix. Then find and print the maximum value, the minimum value, and their respective row and column positions.

Example output:
```
Maximum: 99 at position (row=3, col=1)
Minimum: 2 at position (row=0, col=4)
```

````python
# Write your solution here
````

````python
# SOLUTION
import numpy as np

np.random.seed(42)  # For reproducibility
matrix = np.random.randint(1, 101, size=(5, 5))
print("Matrix:\n", matrix)

max_val = matrix.max()
max_pos = np.unravel_index(matrix.argmax(), matrix.shape)
min_val = matrix.min()
min_pos = np.unravel_index(matrix.argmin(), matrix.shape)

print(f"Maximum: {max_val} at position (row={max_pos[0]}, col={max_pos[1]})")
print(f"Minimum: {min_val} at position (row={min_pos[0]}, col={min_pos[1]})")
````

## Problem 8: NumPy Array Operations

## Problem 9: Boolean Indexing on Exam Scores
Given a NumPy array of exam scores, use boolean indexing to select and print only the passing scores (scores greater than or equal to 60). Also print the number of passing students and the number of failing students.

Use the following scores:
```python
scores = np.array([45, 78, 92, 33, 60, 55, 88, 71, 49, 95])
```

Expected output:
```
Passing scores: [78 92 60 88 71 95]
Passing: 6, Failing: 4
```

````python
# Write your solution here
````

````python
# SOLUTION
import numpy as np

scores = np.array([45, 78, 92, 33, 60, 55, 88, 71, 49, 95])

passing = scores[scores >= 60]
num_passing = len(passing)
num_failing = len(scores) - num_passing

print("Passing scores:", passing)
print(f"Passing: {num_passing}, Failing: {num_failing}")
````
