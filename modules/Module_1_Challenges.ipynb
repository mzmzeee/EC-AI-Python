## Problem 3: Split Email Addresses into Username and Domain

Write a program that reads a list of email addresses and produces a list of `(username, domain)` tuples. Use the `split()` method to separate each address at the `'@'` character.

**Example:**
```
Input:  ["alice@example.com", "bob@college.edu", "charlie@domain.org"]
Output: [("alice", "example.com"), ("bob", "college.edu"), ("charlie", "domain.org")]
```

**Requirement:** Do not hard-code the indices; use `split()` to locate the `'@'` symbol dynamically.

````python
# Write your solution here
````

````python
# SOLUTION
emails = ["alice@example.com", "bob@college.edu", "charlie@domain.org"]

result = []
for email in emails:
    username, domain = email.split('@')
    result.append((username, domain))

print("Input emails:", emails)
print("Username-domain tuples:", result)
````

## Problem 6: Separate Positive and Negative Numbers

Write a program that takes a list of integers and separates them into two lists: one containing all positive numbers and one containing all negative numbers. Ignore zero (do not include it in either list). Print both resulting lists.

**Example:**
```
Input:  [3, -1, 4, -5, 2, -8, 7, -3, 0]
Positives: [3, 4, 2, 7]
Negatives: [-1, -5, -8, -3]
```

````python
# Write your solution here
````

````python
# SOLUTION
numbers = [3, -1, 4, -5, 2, -8, 7, -3, 0]

positives = []
negatives = []

for n in numbers:
    if n > 0:
        positives.append(n)
    elif n < 0:
        negatives.append(n)

print("Original list:", numbers)
print("Positive numbers:", positives)
print("Negative numbers:", negatives)
````

## Problem 7: Alternate Case Conversion by Index

Write a program that reads a string and applies alternating case: characters at even indices are converted to uppercase, and characters at odd indices are converted to lowercase. Spaces and punctuation remain unchanged in position but follow the same index-based rule.

**Example:**
```
Input:  "python programming"
Output: "PyThOn pRoGrAmMiNg"
```

**Requirement:** Use a loop with index tracking (e.g., `enumerate()`).

````python
# Write your solution here
````

````python
# SOLUTION
text = "python programming"

result = ""
for idx, char in enumerate(text):
    if idx % 2 == 0:
        result += char.upper()
    else:
        result += char.lower()

print("Original:", text)
print("Alternating case:", result)
````

## Problem 8: Character Frequency Analysis

Write a program that counts the frequency of each character in a given string. Store the counts in a dictionary and display them in alphabetical order by character.

**Example:**
```
Input:  "hello world"
Output:
  ' ' : 1
  'd' : 1
  'e' : 1
  'h' : 1
  'l' : 3
  'o' : 2
  'r' : 1
  'w' : 1
```

**Requirement:** Build the dictionary using `dict.get()` or an equivalent membership test.

````python
# Write your solution here
````

````python
# SOLUTION
text = "hello world"

frequency = {}
for ch in text:
    frequency[ch] = frequency.get(ch, 0) + 1

print("Character frequency in '{}':".format(text))
for ch in sorted(frequency):
    print("  '{}' : {}".format(ch, frequency[ch]))
````

## Problem 9: Find the Most Frequent Character

Write a program that finds the character that appears most frequently in a string, ignoring spaces. If multiple characters share the maximum count, return any one of them.

**Example:**
```
Input:  "hello world python programming"
Output: Most frequent character: 'o' (appears 4 times)
```

````python
# Write your solution here
````

````python
# SOLUTION
text = "hello world python programming"

# Remove spaces
filtered = text.replace(" ", "")

# Build frequency dictionary
frequency = {}
for ch in filtered:
    frequency[ch] = frequency.get(ch, 0) + 1

# Find the character with the maximum count
most_frequent = max(frequency, key=frequency.get)

print("Text:", repr(text))
print("Most frequent character:", repr(most_frequent),
      "- appears", frequency[most_frequent], "times")
````

## Problem 10: Sum of a Series

Write a program that computes the sum of the following series for a positive integer `n` entered by the user:

$$S = \frac{1}{1^3} + \frac{1}{2^3} + \frac{1}{3^3} + \dots + \frac{1}{n^3}$$

**Example:**
```
Enter a positive integer n: 3
Sum of the series for n = 3 is: 1.162037037037037
```

**Requirement:** Use a `for` loop. Do not use the `math` module.

````python
# SOLUTION
# n = int(input("Enter a positive integer n: "))  # interactive version
n = 5  # default value for execution

total = 0.0
for i in range(1, n + 1):
    total += 1 / (i ** 3)

print("Sum of the series for n =", n, "is:", total)
````

## Problem 11: Print a Diamond Pattern

Write a program that prints a diamond (rhombus) pattern of asterisks using nested loops. The pattern is symmetrical about a horizontal centre line. Use a variable `n` to control the number of rows in the upper half (including the centre row).

**Example for n = 4:**
```
   *
  ***
 *****
*******
 *****
  ***
   *
```

**Requirement:** Use two separate loops: one for the upper half (growing rows) and one for the lower half (shrinking rows). Use string repetition (`*`) rather than an inner printing loop for the asterisks.

````python
# Write your solution here
````

````python
# SOLUTION
n = 4   # Number of rows in the upper half (including the centre row)

# Upper half: increasing number of stars
for i in range(1, n + 1):
    spaces = " " * (n - i)
    stars = "*" * (2 * i - 1)
    print(spaces + stars)

# Lower half: decreasing number of stars
for i in range(n - 1, 0, -1):
    spaces = " " * (n - i)
    stars = "*" * (2 * i - 1)
    print(spaces + stars)
````

## Problem 12: Combine Two Lists Using zip()

Write a program that defines two lists — one containing product names and another containing their corresponding prices — and uses the built-in `zip()` function to combine them into a list of tuples. Print each `(product, price)` pair on its own line.

**Example:**
```
Products: ['Laptop', 'Mouse', 'Keyboard', 'Monitor']
Prices:   [1200.50, 25.99, 79.99, 350.00]
Output:
Laptop: $1200.50
Mouse: $25.99
Keyboard: $79.99
Monitor: $350.00
```

**Requirement:** Use `zip()` to iterate over both lists simultaneously. Do not use index-based access.

````python
# Write your solution here
````

````python
# SOLUTION
products = ['Laptop', 'Mouse', 'Keyboard', 'Monitor']
prices = [1200.50, 25.99, 79.99, 350.00]

# Use zip() to iterate over both lists in parallel
for product, price in zip(products, prices):
    print(f"{product}: ${price:.2f}")
````

## Problem 13: Sort a List Using the key Parameter

Write a program that defines a list of strings representing words with mixed case and sorts them in two different ways using the `sort()` method with the `key` parameter:

1. Sort the list in **case-insensitive alphabetical order**.
2. Sort the list by **string length** (shortest to longest). If two strings have the same length, their original relative order should be preserved.

Print the list after each sorting operation.

**Example:**
```
Original: ['python', 'Java', 'C', 'Rust', 'javascript', 'Go']

After case-insensitive sort:
['C', 'Go', 'Java', 'javascript', 'python', 'Rust']

After sorting by length:
['C', 'Go', 'Java', 'Rust', 'python', 'javascript']
```

**Requirement:** Use `sort(key=...)` for both operations. Do not use `sorted()`.

````python
# Write your solution here
````

````python
# SOLUTION
words = ['python', 'Java', 'C', 'Rust', 'javascript', 'Go']

print("Original:", words)

# Sort case-insensitively by converting each word to lowercase for comparison
words.sort(key=str.lower)
print("\nAfter case-insensitive sort:")
print(words)

# Sort by string length; Python's sort is stable, so equal-length strings
# retain their previous relative order.
words.sort(key=len)
print("\nAfter sorting by length:")
print(words)
````

## Problem 14: Solve a Geometry Problem Using the math Module

Write a program that imports the `math` module and computes the area of a circle and the volume of a sphere for a radius value entered by the user. Use `math.pi` for the constant pi. Print both results rounded to three decimal places.

Formulae:
- Area of a circle: $A = \pi r^2$
- Volume of a sphere: $V = \frac{4}{3} \pi r^3$

**Example:**
```
Enter a radius: 5
Area of circle with radius 5: 78.540
Volume of sphere with radius 5: 523.599
```

**Requirement:** Use `import math` and access `math.pi`. Do not hard-code the value of pi.

````python
# Write your solution here
````

````python
# SOLUTION
import math

# radius = float(input("Enter a radius: "))  # interactive version
radius = 5.0  # default value for execution

# Compute area of a circle
circle_area = math.pi * (radius ** 2)

# Compute volume of a sphere
sphere_volume = (4 / 3) * math.pi * (radius ** 3)

print(f"Area of circle with radius {radius}: {circle_area:.3f}")
print(f"Volume of sphere with radius {radius}: {sphere_volume:.3f}")
````
